```python
import pandas as pd
```

```python
data_import = pd.read_csv("data_cleaning_challenge.csv")
data_import
```

```python
# drop columns with no data
drop_extra = data_import.drop(columns=["Unnamed: 9", "Unnamed: 10"])
drop_extra
```

```python
drop_nan = drop_extra[drop_extra["Row Type"].notna()]
drop_nan
```

```python
# create list to assign number value to each person
column_values = []
counter = 0

for i in drop_nan["Row Type"]:
    if "first name" in i:
        counter += 1
    column_values.append(counter)
```

```python
iter_cols = drop_nan
iter_cols["Iteration"] = column_values
iter_cols
```

```python
# drop extra index 2
drop_col_name_rows = iter_cols[iter_cols["Row Type"] != "Row Type"]
drop_col_name_rows
```

```python
name_df = drop_col_name_rows[drop_col_name_rows["Row Type"].str.contains("first name")]
name_df.drop(columns=["Speed1", "Speed2", "Electricity", "Effort", "Weight", "Torque"], inplace=True)
name_df.rename(columns={"Row Type": "First Name", "Iter Number": "Last Name", "Power1": "Date"}, inplace=True)
name_df['First Name'] = name_df['First Name'].str[12:]
name_df['Last Name'] = name_df['Last Name'].str[11:]
name_df['Date'] = name_df['Date'].str[6:]
name_df
```

```python
# use from above, but use inverse operator
no_name_df = drop_col_name_rows[~drop_col_name_rows["Row Type"].str.contains("first name")]
no_name_df
```

```python
# merge dfs - can drop iteration column
clean_df=pd.merge(left=name_df, right=no_name_df, how="inner", on="Iteration")
clean_df.drop(columns=["Iteration"], inplace=True)
```

```python
# export as csv
clean_df.to_csv("cleaned_data.csv", sep=",", encoding="utf-8", index=False)
```
